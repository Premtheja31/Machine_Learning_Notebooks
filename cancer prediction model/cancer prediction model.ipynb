```python
import numpy as np
import pandas as pd
import seaborn as sns
import dtale
```

```python
df=pd.read_excel("cancer patient data sets.xlsx")
print(df.shape)
df.head()
```

```python
df.nunique()
```

```python
scale_mapper = {"Low":1, "Medium":2, "High":3}
df["Level"] = df["Level"].replace(scale_mapper)
```

```python
df.drop(labels=["Patient Id"],axis=1,inplace=True)
```

```python
dtale.show(df)
```

```python
s = df['Age']
q1 = s.quantile(0.25)
q3 = s.quantile(0.75)
iqr = q3 - q1
iqr_lower = q1 - 1.5 * iqr
iqr_upper = q3 + 1.5 * iqr
ad=[x for x in df["Age"] if x>=iqr_lower and x<=iqr_upper]
df=df[df["Age"].isin(ad)]
```

```python
df.shape
```

```python
data_no_corr=df
data_no_corr.drop(labels=["Age","Gender"],axis=1,inplace=True)
```

```python
data_above_60_corr=df[["Air Pollution","Alcohol use","Dust Allergy","OccuPational Hazards","Genetic Risk","chronic Lung Disease","Balanced Diet","Obesity","Passive Smoker","Chest Pain","Coughing of Blood","Fatigue","Level"]]
print(data_above_60_corr.shape)
```

```python
data_above_70_corr=df[["Alcohol use","Dust Allergy","Genetic Risk","Balanced Diet","Obesity","Passive Smoker","Coughing of Blood","Level"]]
print(data_above_70_corr.shape)
```

```python
data_above_75_corr=df[["Obesity","Coughing of Blood","Level"]]
print(data_above_75_corr.shape)
```

# model with total data

```python
from sklearn.model_selection import train_test_split
Xdf_train,Xdf_test,Ydf_train,Ydf_test=train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.25,random_state=0)
```

```python
from flaml import AutoML
automl=AutoML()
```

```python
automl_settings = {
    "time_budget": 100,
    "metric": 'accuracy',
    "task": 'classification',
    "log_file_name": "cancer_prediction_log_file.log",
}
automl.fit(X_train=Xdf_train, y_train=Ydf_train,
           **automl_settings)
```

```python
Y_pred=automl.predict(Xdf_test)
Y_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
print(accuracy_score(Y_pred,Ydf_test))
```

```python
print(classification_report(Y_pred,Ydf_test))
```

```python
details_report=pd.DataFrame({"Level":["Low","Medium","High"],"Given no":[1,2,3]})
details_report
```

Here with using all the data we got 100% accuracy of our Cancer dataset
